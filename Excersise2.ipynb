```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# exercise A
train_data = np.loadtxt('train.txt');
test_data = np.loadtxt('test.txt');

xtrain = np.float32(train_data[0]);
ytrain = np.float32(train_data[1]);
xtest = np.float32(test_data[0]);
ytest = np.float32(test_data[1]);
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)
ax1.set_title('Train data')
ax1.plot(xtrain,ytrain)
ax2.set_title('Test data')
ax2.plot(xtest,ytest)
```

```python
def relu(x):
    return max(0,x)
y_relu = []
y_relu_test = []
for i in range(len(xtrain)):
    y_relu.append(relu(ytrain[i]))
for i in range(len(xtest)):
    y_relu_test.append(relu(ytest[i]))

xtrain = np.float32(train_data[0]);
ytrain = np.float32(train_data[1]);
xtest = np.float32(test_data[0]);
ytest = np.float32(test_data[1]);
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)
ax1.set_title('Train data')
ax1.plot(xtrain,y_relu)
ax1.plot(xtrain,ytrain)
ax2.set_title('Test data')
ax2.plot(xtest,y_relu_test)
ax2.plot(xtest,ytest)
```

```python

##Exercise B
# Train a model 
model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation=tf.nn.relu, input_shape=[1]),
    tf.keras.layers.Dense(32, activation=tf.nn.relu),
    tf.keras.layers.Dense(32, activation=tf.nn.relu),
    tf.keras.layers.Dense(32, activation=tf.nn.relu),
    tf.keras.layers.Dense(32, activation=tf.nn.relu),
    tf.keras.layers.Dense(32, activation=tf.nn.relu),
    tf.keras.layers.Dense(32, activation=tf.nn.relu),
    tf.keras.layers.Dense(32, activation=tf.nn.relu),
    tf.keras.layers.Dense(1)
  ])
prediction_untrained = model.predict(xtest)

checkpoint_path = "training_2/cp.ckpt"

cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                                 save_weights_only=True,
                                                 verbose=1)
# Fit a model
optimizer = tf.keras.optimizers.RMSprop(0.01)
model.compile(loss='mean_squared_error',optimizer=optimizer)
history = model.fit(xtest, ytest,verbose=1, epochs=1000, 
                    validation_data=(xtest, ytest), callbacks=[cp_callback])
# history = model.fit(xtest, ytest, epochs=1000)
```

```python
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)
ax1.set_title('Test data')
ax2.set_title('Untrained prediction')
ax1.plot(xtest,ytest)
ax2.plot(xtest,prediction_untrained)
plt.savefig('Prediction_untrained_2.png')
```

```python
y_pred_train = model.predict(xtrain)
y_pred_test = model.predict(xtest)
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)
ax1.set_title('Train data')
ax2.set_title('Test data')
ax1.plot(xtrain,ytrain)
ax1.plot(xtrain,y_pred_train)
ax2.plot(xtest,ytest)
ax2.plot(xtest,y_pred_test)
ax2.legend(['Test data', 'Trained prediction'], loc='upper left')
ax1.legend(['Train data', 'Trained prediction'], loc='upper left')
ax1.set_xlabel('step')
plt.savefig('Prediction_exc2.png')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Train and validation losses')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train loss', 'Validation loss'], loc='upper right')
plt.savefig('Train_and_val_losses_2.png')
plt.show()
```
